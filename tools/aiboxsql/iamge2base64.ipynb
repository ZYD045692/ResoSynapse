```python
import base64

def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        base64_data = base64.b64encode(image_file.read())
    return base64_data.decode("utf-8")  #  Base64 [^2^][^4^][^5^]

# 
image_path = "base64\9540.jpg_wh300.jpg"
base64_string = image_to_base64(image_path)
print(base64_string)
```

```python
import os
src_dir = 'base64'
pdf_str=""
for i,image_path in enumerate(os.listdir(src_dir)):
    image_path = os.path.join(src_dir,image_path)
    base64_string = image_to_base64(image_path)
    image_str = f"""
    {{
        id: {i+1},
        positionIndex: {i+1},
        data:`data:image/png;base64,{base64_string}`
    }},"""
    pdf_str += image_str
```

```python
pdf_str_ = f"""import type {{PdfServiceDetailType }}  from "../../api/model/pdfServiceDetailModel";
export const PdfServiceDetailList: PdfServiceDetailType[] = [{pdf_str}];"""
# print(pdf_str_)
```

```python
with open('pdfServiceDatailList.ts','w') as file:
    file.write(pdf_str_)
```
