```python
import requests
import os
import re
from bs4 import BeautifulSoup
src_url = 'https://yanweb.top/'
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36 Edg/133.0.0.0'
response = requests.get(src_url,headers={'User-Agent':user_agent})
soup = BeautifulSoup(response.text, 'html.parser')
groups_element = soup.find_all('ul', class_="mylist row")
```

```python
from bs4 import BeautifulSoup

def format_svg(svg_content):
    # 使用 BeautifulSoup 解析 SVG
    soup = BeautifulSoup(svg_content, 'html.parser')
    svg_element = soup.find('svg')
    
    # 调整属性顺序和格式
    svg_element.attrs = {
        'version': svg_element.get('version', '1.1'),
        'viewBox': svg_element.get('viewbox', '0 0 1024 1024'),
        'class': svg_element.get('class', 'icon'),
        'xmlns': 'http://www.w3.org/2000/svg',
        'width': svg_element.get('width', '200'),
        'height': svg_element.get('height', '200'),
        't': svg_element.get('t', '1738320968669'),  # 新增的 't' 属性
        'p-id': svg_element.get('p-id', '6337'),    # 新增的 'p-id' 属性
    }

    # 清理 SVG 代码，去掉冗余空格和换行
    formatted_svg = soup.prettify(formatter=lambda s: s if s else ' ')
    return formatted_svg

def save_svg_file(formatted_svg, filename):
    # 保存为 .svg 文件
    with open(filename, 'w') as f:
        f.write(formatted_svg)

# 示例输入 SVG 代码
input_svg = """
<svg class="icon" height="200" p-id="6337" t="1738320968669" version="1.1" viewbox="0 0 1024 1024" width="200" xmlns="<url id="cv2nat8nnlr42t1766o0" type="url" status="parsed" title="SVG namespace" wc="777">http://www.w3.org/2000/svg</url>   "><path d="M315.584 277.802667a41.813333 41.813333 0 1 1-83.605333 0.021333 41.813333 41.813333 0 0 1 83.605333-0.021333z m476.458667 0a41.813333 41.813333 0 1 1-83.584 0.021333 41.813333 41.813333 0 0 1 83.584-0.021333zM231.957333 746.197333a41.813333 41.813333 0 1 1 83.605334-0.042666 41.813333 41.813333 0 0 1-83.605334 0.042666z m476.48 0a41.813333 41.813333 0 1 1 83.584-0.042666 41.813333 41.813333 0 0 1-83.584 0.042666zM441.258667 361.962667a73.088 73.088 0 1 0 0 146.176 73.088 73.088 0 0 0 0-146.176z m310.528 124.202666H853.333333v51.690667h-101.546666a241.237333 241.237333 0 0 1-213.952 213.909333v101.546667h-51.669334v-101.546667a241.237333 241.237333 0 0 1-213.930666-213.909333H170.666667v-51.690667h101.568A241.237333 241.237333 0 0 1 486.165333 272.213333V170.666667h51.669334v101.546666A241.216 241.216 0 0 1 751.786667 486.186667z" fill="#4C4C4C" p-id="6338"></path></svg>
"""

# 格式化 SVG
formatted_svg = format_svg(input_svg)

# 保存为文件
save_svg_file(formatted_svg, 'standardized_icon.svg')

print("SVG 文件已保存为 'standardized_icon.svg'")
```

```python
save_path = './images'
serve_path = 'images/'
```

```python
def save_svgandpng(svgorimg,name,save_path = save_path,serve_path = serve_path):
    file_path = f"{save_path}/{name}"
    if isinstance(svgorimg,bytes):
        file_path += '.png'
        with open(file_path, "wb") as f:
            f.write(svgorimg)
        return os.path.join(serve_path,f'{name}.png')
    
    else:
        file_path += '.svg'
        svg = format_svg(str(svgorimg))
        save_svg_file(svg, file_path)
        return os.path.join(serve_path,f'{name}.svg')
```

```python
# CREATE TABLE major_widgets (
#     widget_id INT PRIMARY KEY AUTO_INCREMENT,
# major_name VARCHAR(50) NOT NULL,
# icon_url VARCHAR(200),
#     display_order INT DEFAULT 0,
#     is_visible BOOLEAN DEFAULT TRUE
# );
```

```python
major_widget_id = []
major_icon_url = []
major_major_name = []
major_display_order = []
major_dict = dict()
for group in  groups_element:
    widget_id = len(major_widget_id) + 1
    display = len(major_display_order) + 1
    icon = group.find_all('svg',class_="icon")[0]
    major_name = group.find('sapn').text
    icon_url = save_svgandpng(icon,major_name)

    major_widget_id.append(widget_id)
    major_display_order.append(display)
    major_icon_url.append(icon_url)
    major_major_name.append(major_name)
    major_dict[major_name] = widget_id
```

```python
widget_assign_major_widget_id = []
widget_assign_minor_widget_id = []

minor_widget_id = []
minor_name = []
minor_widget_url = []
minor_icon_url = []
minor_display_order = []
for group in  groups_element:
    for i,minor in enumerate(group.find_all('a')):
        minor_name.append(minor.text)
        minor_widget_id.append(len(minor_widget_id) + 1)
        minor_widget_url.append(minor['href'].replace('yanweb',''))
        minor_display_order.append(i+1)

        widget_assign_major_widget_id.append(major_dict[group.find('sapn').text])
        widget_assign_minor_widget_id.append(minor_widget_id[-1])

        if minor.find('img'):
            img = minor.find('img')
            img_url = os.path.join(src_url,(re.findall(r'src="(.*?)"', str(minor.find('img'))))[0])
            img = requests.get(img_url,headers={'User-Agent':user_agent}).content
            icon_url = save_svgandpng(img,minor.text)
        else:
           svg = minor.find('svg')
           icon_url = save_svgandpng(svg,minor.text)
        minor_icon_url.append(icon_url)
```
